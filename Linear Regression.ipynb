### Gradient Algorithm
    - Start with a random value of theta
    - Repeat a task till convergence
        - Update theta as per rule

```python
def hypothesis(X,theta):                             # function for calculating hypothesis
    return theta[0] + theta[1]*X

def error(X,Y,theta):
    n = X.shape[0]                                   # function for calculating error
    total_error = 0
    for i in range(n):
        total_error += (Y[i]-hypothesis(X[i],theta))**2
    
    return 0.5*total_error

def gradient(X,Y,theta):
    
    grad = np.zeros((2,))
    n = X.shape[0]
    for i in range(0,n):                             # function returns the value of gradient 
        hx = hypothesis(X[i],theta)
        grad[0] += (hx-Y[i])
        grad[1] += (hx-Y[i])*X[i]
    
    return grad

def GradientDescentRule(X,Y,learning_rate=0.001):
    theta = np.zeros((2,))     # starting with random value of theta
    i = 0
    e = []
    while(i<=100):
        grad = gradient(X,Y,theta)
        theta[0] = theta[0] - grad[0]*learning_rate
        theta[1] = theta[1] - grad[1]*learning_rate
        e.append(error(X,Y,theta))
        #if i>=1:
            #if abs(e[len(e)-1]-e[len(e)-2])<=0.001:
                #break
        i += 1
        
    return theta,e

    
```

```python
theta,err = GradientDescentRule(X,Y)
a = np.linspace(-2.5,3,10)
print(a)
```

```python
y = hypothesis(a,theta)
print(y)
```

```python
plt.scatter(X,Y,color='blue',label='training data')
plt.plot(a,y,color='orange',label='test data')
plt.legend()
plt.show()
```
