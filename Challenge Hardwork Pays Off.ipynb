```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dx = pd.read_csv('Linear_X_Train.csv')
dy = pd.read_csv('Linear_Y_Train.csv')
X = dx.values
Y = dy.values
print(X.shape,Y.shape)
#print(X)
#Y = (Y - Y.mean())/Y.std()
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
def hypothesis(x,theta):
    return theta[0] + theta[1]*x

def error(X,Y,theta):
    n = X.shape[0]
    total_error = 0
    for i in range(n):
        total_error += (hypothesis(X[i],theta) - Y[i])**2
        
    return 0.5*total_error


def gradient(X,Y,theta):
    
    grad = np.zeros((2,))
    n = X.shape[0]
    for i in range(n):
        hx = hypothesis(X[i],theta)
        grad[0] += (hx - Y[i])
        grad[1] += (hx - Y[i])*X[i]
        
    return grad
        
    
def GradientDescent(X,Y,learning_rate=0.0001):
    theta = np.zeros((2,))
    e = []
    i=0
    while(i<=500):
        grad = gradient(X,Y,theta)
        theta[0] = theta[0] - learning_rate*grad[0]
        theta[1] = theta[1] - learning_rate*grad[1]
        e.append(error(X,Y,theta))
        i+=1
        
    return theta,e

```

```python
theta,err = GradientDescent(X,Y)
print(theta)
a = np.linspace(-3,4,15)
plt.scatter(X,Y)
plt.plot(a,hypothesis(a,theta),color='red')
plt.show()
#print(a)
plt.plot(err)
```
